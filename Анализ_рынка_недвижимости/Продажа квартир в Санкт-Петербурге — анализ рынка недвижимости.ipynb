```python
df['ceiling_height'].describe()
```

```python
df['ceiling_height'].hist()
plt.title("гистограмма по параметру ceiling_height")
plt.xlabel("высота")
plt.ylabel("количество")
plt.show()
```

На графике видно, что большее количество данных встречается в районе от 1 до 10 метров, но присутствуею данные со значениями и 100 метров.
<br>Минимальное и максимальное значения в столбце `ceiling_height` не правдоподобны. Пройдемся по минимальным значениям в столбце. Потолки равные 1 м в жилых помещениях - очень сомнительно. Заменим их на среднее значение. <br> Допустим что высота потолков в 6 метров приемлема для двууровневых квартир, но вот высота  выше 6 м вызывает вопросы. Изучим высоту более 6 метров и заменим.
<br> Проведем корректировку данных в 2 этапа: 
*  Данные, распологающиеся от 20 до 100, разделим на 10, так как высока вероятность, что точка установлена не правильно и значение 25 м. на самом деле равно 2.5 метрам.
*  Данные расположенные до 2 метров, от 6 до 20 метров и 100 метров, заменим медиальными значениями.
* Значения Nan так же заменим медианами.

```python
df[(df['ceiling_height']>=20) & (df['ceiling_height']<100)] 
#просмотр строк с высотой потолков от 20 до 100
```

```python
df.loc[((df['ceiling_height'] >= 20) & (df['ceiling_height'] <  100)), 'ceiling_height'] /= 10
#замена значений от 20 до 100 в столбце ceiling_height
```

```python
df.loc[((df['ceiling_height'] > 6) & (df['ceiling_height'] < 20)), 'ceiling_height'] = df.groupby(['locality_name'])['ceiling_height'].transform('median')
#перезапись сзначений от 6 до 20 медианой групировки по локации
df.loc[((df['ceiling_height'] >= 1) & (df['ceiling_height'] < 2)), 'ceiling_height'] = df.groupby(['locality_name'])['ceiling_height'].transform('median')
#перезапись сзначений от 1 включительно до 2 медианой групировки по локации
df.loc[(df['ceiling_height'] >= 100 ), 'ceiling_height'] = df.groupby(['locality_name'])['ceiling_height'].transform('median')
#замена значений от 100 и выше медианами в зависимости от локации
```

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby(['locality_name'])['ceiling_height'].transform('median'))
#заполнение оставшихся Nan медианами с групировкой по количеству комнат
```

У нас остались строки с пропущенными значениеми в столбце `ceiling_height`. Посмотрим на них:

```python
df[df['ceiling_height'].isna()]
```

Т.к. мы использовали для вычисления медианы группировку по локации, строки в которых локации пропущены или локация повторяется всего один раз - не заполнились. <br>Дозаполним пропуски медианой по группировке от количества комнат.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby(['rooms'])['ceiling_height'].transform('median'))
#заполнение оставшихся Nan медианами с групировкой по количеству комнат
df['ceiling_height'].describe()
```

```python
df['ceiling_height'].hist()
plt.title("гистограмма по параметру ceiling_height")
plt.xlabel("высота")
plt.ylabel("количество")
plt.show()
```

`Nan`ы в столбце `ceiling_height`заполнены. Для того, что бы в дальнейшем было меньше ошибок при заполнении данного поля, можно сделать его обязательным для заполнения и возможно внести разумные лимиты для минимальных и максимальных значений.<br> Перейдем к пропускам в столбце **`floors_total`** - количество этажей в доме.

```python
df['floors_total'].hist()
plt.title("гистограмма по параметру floors_total")
plt.xlabel("количество этажей")
plt.ylabel("количество")
plt.show()
```

```python
df['floors_total'].describe() #вывод статистике по столбцу floors_total
```

График показывает, а статиска это подтверждает, что максимальное значение этажности равно 60. Рассмотрим подробнее значение этажности выше 36 (интернет говорит, что это максимальный этаж ЖК в СПб). 

```python
df[df['floors_total'] > 36 ]
```

Воспользуемся схемой  `ceiling_height` для выбросов в `floors_total`.

```python
df.loc[(df['floors_total'] > 36) & (df.groupby(['locality_name'])['floors_total'].transform('median')<df['floor']), 'floors_total'] = df['floor']
#перезапись значений выше 36, при условии, что медиана с группировкой по локации  меньше этажа
#квартиры, на этаж, на котором расположена квартира.
df.loc[(df['floors_total'] > 36) & (df.groupby(['locality_name'])['floors_total'].transform('median')>= df['floor']), 'floors_total'] = df.groupby(['locality_name'])['floors_total'].transform('median')
#перезапись значений выше 36, при условии, что медиана с группировкой по локации  больше этажа
#квартиры, на медиану с группировкой по локации
#df[(df['locality_name'] == 'Санкт-Петербург') & (df['total_area'] == 40)& (df['last_price'] == 3978000.0)]
```

Заменим значения `Nan` в столбце `floors_total` на медиальные значения по названию населеного пункта, если медиана меньше, чем значение в столбце `floor`, то в столбцец `floors_total` подставим значение из `floor`.

```python
df.loc[(df['floors_total'].isnull()) & (df.groupby(['locality_name'])['floors_total'].transform('median')>= df['floor']), 'floors_total'] = df.groupby(['locality_name'])['floors_total'].transform('median')
# замена Nan  по условию если медиана больше этажа квартиры, которая продается в группировке по локации
df.loc[(df['floors_total'].isnull()) & (df.groupby(['locality_name'])['floors_total'].transform('median')<df['floor']), 'floors_total'] = df['floor']
# замена Nan  по условию если медиана меньше этажа квартиры, которая продается в группировке по локации
df.loc[(df['floors_total'].isnull()) & (df.groupby(['rooms'])['floors_total'].transform('median')>= df['floor']), 'floors_total'] = df.groupby(['rooms'])['floors_total'].transform('median')
# замена Nan  по условию если медиана больше этажа квартиры, которая продается в группировке по количеству комнат
```

```python
df['floors_total'].hist()
plt.title("гистограмма по параметру floors_total")
plt.xlabel("количество этажей")
plt.ylabel("количество")
plt.show()
```

```python
df['floors_total'].describe()#вывод статистики после замены Nan
```

На графике видно, что максимальным значением сейчас считается 36 этажей. Основной пик данных приходится на 5-7 этажей. <br> Заменим тип данных на `int`, т.к. количество этажей - это целые числа.

```python
df['floors_total'] = df['floors_total'].astype('int') # замена типа данных столбца floors_total
```

Данные по столбцу `floors_total` заполнены, как и для столбца `ceiling_height`, возможно, правильно было бы при заполнении формы сделать этот параметр обязательным. <br> Приступим к рассмотрению следующего столбца **`balcony`**.

К сожалению, у большинства квартир, балконы отсутствуют. <br> Приступим к обработке значений в столбцах : `kitchen_area` и `living_area`.<br> Если квартира является квартирой-студией, то площадь кухни равна 0, т.к. она входит в жилую площадь.
```python
df.loc[(df['kitchen_area'].isnull())&(df['studio'] == True), 'kitchen_area'] = 0

#Замена Nan в столбце kitchen_area, если квартира-студия.
```

Заменим `Nan`  в столбцах `kitchen_area` и  `living_area` по пропорции с `total_area` по группировке по количеству комнат в квартире.

```python
df.loc[df['living_area'].isnull(), 'living_area']  = round((df.groupby(['rooms'])['living_area'].transform('median') / df.groupby(['rooms'])['total_area'].transform('median') * df['total_area']),2)
#замена nan в столбце living_area пропорцией
df.loc[df['kitchen_area'].isnull(), 'kitchen_area']  = round((df.groupby(['rooms'])['kitchen_area'].transform('median') / df.groupby(['rooms'])['total_area'].transform('median') * df['total_area']),2)
#замена nan в столбце kitchen_area пропорцией
```

```python
df[['living_area','kitchen_area']].describe() #вывод статистики по столбцам living_area иkitchen_area
```

```python
df[df['living_area'] > df['total_area']] #Проверка того, что,жилая площадь не привышает общую
```

Мы заменили `Nan` в столбцах `living_area` и `kitchen_area`. <br> У нас осталось довольно много столбцов с пропущенными значеними. `Nan`ы в них мы решили не менять:
* для столбцов `locality_name`, `airports_nearest`,`city_сenters_nearest`, `parks_around_3000, parks_nearest`, `ponds_around_3000` и `ponds_nearest` - нет достаточных данных, что бы определить населенный пункт для пропусков (адреса), а приблизительные, средние или медиальные значения, возможно приведут к искажению.

<br> Приведем еще пару столбцов в более удобный вид, отформатируем даты, дни публикации и приведем цены в тысячи:

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')#форматируем столбец с датами
df['last_price'] = df['last_price']/1000 #перевод столбца с ценами в тысячи
df['days_exposition'] = df['days_exposition'].fillna(0).astype('int')

df.info()
```

В предобработке данных мы выявили и заменили значения `Nan` в столбцах `ceiling_height`, `floors_total`, `living_area`,`is_apartment`,`kitchen_area`,`balcony`мы провели операции по корректировке неправдоподобных данных и замене `Nan`-значений. Что занимательно, изменения мы внесли только в те столбцы, которые были заполнены в ручную клиентами.
В оставшихся столбцах с пропусками,которые получены автоматически на основе картографических данных, мы оставили все как есть, потому что, не имеем дополнительных данных для их заполнения.  <br> Так же мы заменили типы данных в некоторых столбцах. Для `is_apartment` сделали его булевым, так как значения  в нем могут быть только **True** or **False**, `first_day_exposition` сделали датами, количество этажей и балконов привели к типу **Int**. Так же для удобства цена на кратиры сделали в тысячах.<br> Скорее всего пропуски в данных возникли из-за неправильного заполнения форм. Даже значения, которые должны были проставиться автоматически, например,`parks_around_3000`, не заполнились, скорее всего, из-за того, что адрес был указан не верно. Для того, что бы в дальнейшем сократить `Nan` в данных, необходимо поработать с формой, которую заполняют клиенты. Например, прописать максимальные значения высоты потолков для заполнения и сделать обязателным заполнение поле `is_apartment`, а так же создать выпадающие списки.
